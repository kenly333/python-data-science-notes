```python
import numpy as np
import matplotlib.pyplot as plt
```

### 根据一个数组绘图

```python
# 只提供给plot()函数一个列表或数组，matplotlib会认为这是一串Y值(Y向量)，并且自动生成X值(X向量)。
# Python一般是从0开始计数的，所以X向量有和Y向量一样的长度(此处是4)，但是是从0开始.

plt.ylabel('y')
plt.xlabel('x')
plt.plot([1,2,3,4],c='red')

plt.show()
```

### 根据一对数组绘图

```python
# 给plt.plot()函数传递多个序列(元组或列表)，每两个序列是一个X,Y向量对，在图中构成一条曲线
x=[1,2,3,4]
y=[1,4,9,16]

plt.ylabel('y')
plt.xlabel('x')
plt.plot(x,y,c='red')

plt.show()
```

### 根据2对数组绘图

```python
# 同一个图里存在多条曲线
x1=[1,2,3,4]
y1=[1,4,9,16]

x2=[1,2,3,4]
y2=[2,5,20,32]

plt.ylabel('y')
plt.xlabel('x')
plt.plot(x1,y1,x2,y2,c='red',linewidth=2.0)

plt.show()
```

### 根据2对数组绘图

```python
# 同一个图里存在多条曲线
x1=[1,2,3,4]
y1=[1,4,9,16]

x2=[1,2,3,4]
y2=[2,5,20,32]

plt.ylabel('y')
plt.xlabel('x')
plt.plot(x1,y1,c='red',linewidth=2.0)
plt.plot(x2,y2,c='green')

plt.show()
```

### 根据函数绘图

```python
# 同一个图里存在多条曲线
coef=0.2
intercept=1

x1 = np.arange(0, 12)
y1 = (intercept + x1 * coef) / coef

coef=0.02
x2 = np.arange(1, 12)
y2 = np.power((1 + 1/x2),x2)

plt.ylabel('y')
plt.xlabel('x')
plt.plot(x1,y1,c='red')
plt.plot(x2,y2,c='green')

#plt.axis([-1,12,0,20])

plt.show()
```
