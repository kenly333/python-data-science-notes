```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2维直线

```python
intercept = np.random.random([1])
coef = np.random.random([2])

lx = np.arange(0, 12)
ly = (intercept + lx * coef[0]) / coef[1]

plt.ylabel('y')
plt.xlabel('x')
plt.plot(lx, ly, c='yellow')
```

# 2维直线

```python
intercept = np.random.random([1])
coef = np.random.random([2])

lx = np.arange(0, 12)
ly = (-intercept - lx * coef[0]) / coef[1]

plt.ylabel('y')
plt.xlabel('x')
plt.plot(lx, ly, c='red')
```
