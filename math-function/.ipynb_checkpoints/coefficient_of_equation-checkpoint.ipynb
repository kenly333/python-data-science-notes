```python
import numpy as np
import matplotlib.pyplot as plt
```

### 函数定义

```python
def regression_fun(x,coef,intercept):
    return (intercept + x * coef[0]) / coef[1]

def regression_fun_neg(x,coef,intercept):
    return (-intercept - x * coef[0]) / coef[1]
```

### 递增2维直线

```python
intercept = np.random.random([1])
coef = np.random.random([2])

lx = np.arange(0, 12)
ly = regression_fun(lx,coef,intercept)

plt.ylabel('y')
plt.xlabel('x')
plt.plot(lx, ly, c='yellow')
```

### 递减2维直线

```python
intercept = np.random.random([1])
coef = np.random.random([2])

lx = np.arange(0, 12)
ly = regression_fun_neg(lx,coef,intercept)

plt.ylabel('y')
plt.xlabel('x')
plt.plot(lx, ly, c='red')
```
